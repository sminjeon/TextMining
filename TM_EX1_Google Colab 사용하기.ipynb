#참고1- Google Colab 단축키 



[Google Colab 단축키 참고 사이트](https://surfonmedia.tistory.com/1)

[실행 관련 단축키]

1. Ctrl + Enter = 해당 셀을 실행하고 커서를 해당 셀에 두는 경우 (결과 값만 보고자 할 때)

2. Shift + Enter = 해당 셀을 실행하고 커서를 다음 셀로 넘기는 경우 (여러가지 값을 빠르게 출력할 때)

3. Alt + Enter = 해당 셀을 실행하고 셀을 삽입한 후 커서를 삽입한 셀로 넘기는 경우 (다음 작업 공간이 없을 때)

[셀 삽입/삭제 관련 단축키]

1. Ctrl + M A = 코드 셀 위에 삽입

2. Ctrl + M B = 코드 셀 아래 삽입

3. Ctrl + M D = 셀 지우기

4. Ctrl + M Y = 코드 셀로 변경

5. Ctrl + M M = 마크다운 셀로 변경

6. Ctrl + M Z = 실행 취소
